```c++
#include <iostream>
using namespace std;
```

```c++
//a.
void a_fun(int n)
{
    if(n==0)
        return;
    cout << n << endl;
    a_fun(n-1);
    cout << n << endl;
}
```

```c++
a_fun(3)
```

```c++
//b.
void b_fun(int n)
{
    if(n==0)
        return;
    b_fun(n-1);
    cout << n << endl;
    b_fun(n-1);
}
```

```c++
b_fun(3)
```

```c++
//c.
int c_fun(int n)
{
    if(n==1)
        return 0;
    else
        return 1+ c_fun(n/2);
}
```

```c++
c_fun(16)
```

```c++
//d.
void d_fun(int n)
{
    if(n==0)
        return;
    d_fun(n/2);
    cout << n%2;
}
```

```c++
d_fun(7)
```

```c++
d_fun(11)
```
