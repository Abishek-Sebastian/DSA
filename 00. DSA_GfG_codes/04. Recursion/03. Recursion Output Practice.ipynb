```c++
#include <iostream>
using namespace std;
```

```c++
//a.
void a_fun(int n)
{
    if(n==0)
        return;
    cout << n << endl;
    a_fun(n-1);
    cout << n << endl;
}
```

```c++
a_fun(3)
```

```c++
//b.
void b_fun(int n)
{
    if(n==0)
        return;
    b_fun(n-1);
    cout << n << endl;
    b_fun(n-1);
}
```

```c++
b_fun(3)
```
